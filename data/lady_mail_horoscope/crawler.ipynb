##### Готовим первичный набор ссылок и заголовков
Парсинг заголовков и ссылок, находящихся на страницах типа https://lady.mail.ru/rubric/horoscope/?page=

```python
driver = webdriver.Chrome('C:\chromedriver_win32\chromedriver.exe')
```

```python
titles = []
links = []

for n in range(1,93):
    driver.get("https://lady.mail.ru/rubric/horoscope/?page=" + str(n))
    time.sleep(random.random() * 10)

    for i in range(1,11):
        block = driver.find_element_by_css_selector("body > div.layout > div:nth-child(22) > div:nth-child(3) > div > div > div > div > div > div.cols.cols_percent.cols_margin > div > div:nth-child(" + str(i) + ") > div > div")
        title = block.find_element_by_class_name("newsitem__title-inner").text
        link = block.find_element_by_tag_name("a").get_attribute("href")
        titles.append(title)
        links.append(link)
        
# Тут нормально, что ошибка, парсинг всё равно до конца прошел.
# Просто на последней странице оказалось 9 ссылок вместо 10
```

```python
df_one = pd.DataFrame(data=[titles, links], index=['titles', 'links']).T
df_one
```

##### Оставляю статьи, относящиеся к конкретным дням

```python
df_one = pd.read_csv('df_one.csv')
df_two = df_one.loc[df_one.titles.str.contains('Гороскоп на ', na=False) & (~df_one.titles.str.contains(' год', na=False))]
df_one.to_csv('df_two.csv', index=False)
df_two
```

```python
df_three = df_two.copy()
df_three['articles'] = ''
df_three['date'] = ''

for index, row in df_three.iterrows():
    driver.get(row['links'])
    time.sleep(random.random() * 10)
    block = driver.find_element_by_css_selector("body > div.layout > div:nth-child(23) > div.block.block_bg_none.block_collapse_hor > div > div > div > div > div > div.article.article_collapsed.js-module.js-article.js-promo-popup-ref > div.article__text.article__text_lady.js-module.js-article_text.margin_bottom_20.js-mediator-article.m-lazy-item_complete")
    blocks = block.find_elements_by_css_selector("div")
    date = driver.find_element_by_css_selector("body > div.layout > div:nth-child(23) > div:nth-child(1) > div > div > div > div.cols__column.cols__column_small_23.cols__column_small_23_5.cols__column_medium_32.cols__column_large_39 > div > div > div.breadcrumbs.breadcrumbs_article > span > span > span")
    df_three.loc[index, 'articles'] = [el.text for el in blocks]
    df_three.loc[index, 'date'] = date.text
    
# ХЗ, что случилось, но немного не допарсилось, поэтому ниже продолжаю
```

```python
def get_separate_sign(article, sign):
    
    article = [el.strip() for el in article]
    
    signs = [
            'Овен', 'Телец', 'Близнецы', 'Рак', 'Лев', 'Дева',
            'Весы','Скорпион', 'Стрелец', 'Козерог', 'Водолей', 'Рыбы'
            ]
    try:
        ind = article.index(sign)
    except ValueError:
        return ''
    
    forecast = ''
    
    for i in range(ind+1, len(article)):
        if article[i] not in signs:
            if article[i] != '':
                forecast += ' ' + article[i]
        if article[i] in signs:
            break
    
    return forecast
            
for sign in ['Овен', 'Телец', 'Близнецы', 'Рак', 'Лев', 'Дева', 'Весы','Скорпион', 'Стрелец', 'Козерог', 'Водолей', 'Рыбы']:
    df_three[sign] = df_three['articles'].apply(get_separate_sign, sign=sign)
```
