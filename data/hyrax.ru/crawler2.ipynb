```python
from selenium import webdriver
import pandas as pd
import time
import random
```

##### Готовим первичный набор ссылок и заголовков
Парсинг заголовков и ссылок, находящихся на страницах типа https://www.hyrax.ru/archive1996.shtml

```python
driver = webdriver.Chrome('C:\chromedriver_win32\chromedriver.exe')
```

```python
titles = []
links = []

for n in range(1996,2010):
    driver.get("https://www.hyrax.ru/archive" + str(n) + ".shtml")
    time.sleep(random.random() * 5)

    elems = driver.find_elements_by_tag_name("a")
    
    for el in elems:
        if "cgi-bin/bn_html.cgi" in el.get_attribute("href"):
            titles.append(el.text)
            links.append(el.get_attribute("href"))
```

```python
for i, j in enumerate(zip(titles, links)):
    if i % 100 == 0:
        print(j)
```

```python
df_one = pd.DataFrame(data=[titles, links], index=['titles', 'links']).T
df_one
```

```python
df_one.to_csv('df_one_hyrax.csv', index=False)
```

```python
driver.close()
```

##### Теперь проходим по собранным ссылкам

```python
df_one = pd.read_csv('df_one_hyrax.csv')
```

```python
df_three = df_one.copy()
df_three['articles'] = ''
df_three['date'] = df_three['titles'].apply(lambda x: x[-10:])

for index, row in df_three.loc[df_three['articles']==''].iterrows():
    driver.get(row['links'])
#     time.sleep(random.random() * 5)
    block = driver.find_element_by_css_selector("body")
    df_three.loc[index, 'articles'] = block.text.split('\n')
```

```python
driver.close()
```

```python
df_three['len_articles'] = df_three['articles'].apply(lambda x: len(x))
```

```python
def get_separate_sign(article, sign):
    
    article = [el.strip() for el in article]
    
    signs = [
            'ОВЕН (21 марта - 20 апреля)', 'ТЕЛЕЦ (21 апреля - 21 мая)', 'БЛИЗНЕЦЫ (22 мая - 21 июня)',
            'РАК (22 июня - 23 июля)', 'ЛЕВ (24 июля - 23 августа)', 'ДЕВА (24 августа - 23 сентября)',
            'ВЕСЫ (24 сентября - 23 октября)','СКОРПИОН (24 октября - 22 ноября)', 'СТРЕЛЕЦ (23 ноября - 21 декабря)',
            'КОЗЕРОГ (22 декабря - 20 января)', 'ВОДОЛЕЙ (21 января - 19 февраля)', 'РЫБЫ (20 февраля - 20 марта)'
            ]
    try:
        ind = article.index(sign)
    except ValueError:
        return ''
    
    forecast = ''
    
    for i in range(ind+1, len(article)):
        if article[i] not in signs:
            if article[i] != '':
                forecast += ' ' + article[i]
        if article[i] in signs:
            break
    
    return forecast
            
for sign in ['ОВЕН (21 марта - 20 апреля)', 'ТЕЛЕЦ (21 апреля - 21 мая)', 'БЛИЗНЕЦЫ (22 мая - 21 июня)',
            'РАК (22 июня - 23 июля)', 'ЛЕВ (24 июля - 23 августа)', 'ДЕВА (24 августа - 23 сентября)',
            'ВЕСЫ (24 сентября - 23 октября)','СКОРПИОН (24 октября - 22 ноября)', 'СТРЕЛЕЦ (23 ноября - 21 декабря)',
            'КОЗЕРОГ (22 декабря - 20 января)', 'ВОДОЛЕЙ (21 января - 19 февраля)', 'РЫБЫ (20 февраля - 20 марта)']:
    df_three[sign] = df_three['articles'].apply(get_separate_sign, sign=sign)
    
df_three.rename({'ОВЕН (21 марта - 20 апреля)': 'Овен', 'ТЕЛЕЦ (21 апреля - 21 мая)': 'Телец',
                 'БЛИЗНЕЦЫ (22 мая - 21 июня)': 'Близнецы', 'РАК (22 июня - 23 июля)': 'Рак', 
                 'ЛЕВ (24 июля - 23 августа)': 'Лев', 'ДЕВА (24 августа - 23 сентября)': 'Дева',
                 'ВЕСЫ (24 сентября - 23 октября)': 'Весы','СКОРПИОН (24 октября - 22 ноября)': 'Скорпион',
                 'СТРЕЛЕЦ (23 ноября - 21 декабря)': 'Стрелец', 'КОЗЕРОГ (22 декабря - 20 января)': 'Козерог',
                 'ВОДОЛЕЙ (21 января - 19 февраля)': 'Водолей', 'РЫБЫ (20 февраля - 20 марта)': 'Рыбы'}, axis=1, inplace=True)
```

```python
df_three['date'] = df_three.articles.apply(lambda x: ' '.join(x[2].split()[1:4]))
df_three['date'] = df_three['date'].apply(lambda x: x.replace(',',''))
df_three
```
